```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import random
```

```python
dataset,y= make_blobs(n_samples=200,n_features=2,centers=3)
```

```python
x=dataset[:,0]
y=dataset[:,1]
```

```python
k=3
centroids={
    i+1:[random.choice(x),random.choice(y)]
    for i in range (k)
}
```

```python

colorm={1:'r',2:'b',3:'m'}
def cluster(x,y,centroids,colorm):
    grp=plt.figure(figsize=(5,5))
    plt.scatter(x,y,color='k')
    for i in centroids.keys():
        plt.scatter(*centroids[i],color=colorm[i],marker='*',s=100)
    
    plt.xlim(min(x)-2,max(x)+2)
    plt.ylim(min(y)-2,max(y)+2)
    plt.show()
```

```python
cluster(x,y,centroids,colorm)
```

```python
distance=[0]*k
def distcalc(x,y,centroids):
    for i in range(len(centroids)):
        distance[i]=np.sqrt(pow((x-centroids[i+1][0]),2)+pow((y-centroids[i+1][1]),2))
```

```python
def assignment(x,y,centroids):
    min_red=[]
    min_blue=[]
    min_purple=[]
    for i in range(len(x)):
        if(distance[0][i]<distance[1][i]):
            if(distance[0][i]<distance[2][i]):
                min_red.append(i)
            else:
                min_purple.append(i)
        elif (distance[1][i]<distance[2][i]):
            min_blue.append(i)
        else:
            min_purple.append(i)
    return min_red,min_blue,min_purple
```

```python
o=1
def plot_scatter(x,y,colorm,o,red,blue,purple,centroids):
    
    for j in red,blue,purple:
        for k in j:
            plt.scatter(x[k],y[k],color=colorm[o],marker='.')
        o=o+1
    for i in centroids.keys():
        plt.scatter(*centroids[i],color=colorm[i],marker='X',s=100)
    plt.xlim(min(x)-2,max(x)+2)
    plt.ylim(min(y)-2,max(y)+2)
    plt.show() 
```

```python
snomean=1
```

```python
centroidspre=[0]
def cluscenter(x,y,centroids,red,blue,purple,snomean):
    i=0
    for j in red,blue,purple:
        meandisx=[]
        meandisy=[]
        for k in j:
            meandisx.append(x[k])
            meandisy.append(y[k])
        meandisx=np.array(meandisx)
        meandisy=np.array(meandisy)
        centroidspre.append(centroids[snomean])
        centroids[snomean][0]=meandisx.mean()
        centroids[snomean][1]=meandisy.mean()
        snomean=snomean+1
        
```

```python
def k_means(x,y,colorm,o,centroids):
    for i in range(10):
        distcalc(x,y,centroids)
        red,blue,purple = assignment(x,y,centroids)
        plot_scatter(x,y,colorm,o,red,blue,purple,centroids)
        cluscenter(x,y,centroids,red,blue,purple,snomean)
```

```python
k_means(x,y,colorm,o,centroids)
```
